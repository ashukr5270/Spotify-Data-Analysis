```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('tracks.csv')
```

```python
df_tracks.head()
```

```python
# check null values

pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
# 10 least popular songs
sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
# 10 most popular songs
most_pop = df_tracks.sort_values('popularity', ascending = False).head(10)
most_pop
```

```python
# another way to get 10 most popular songs

most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
df_tracks.describe().transpose()
```

```python
df_tracks.set_index("release_date", inplace = True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
# TO CONVERT THE DURATION FROM "MILISECONDS" TO "SECONDS"
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
# CORRELATION MAP
corr_df = df_tracks.drop(["key", "mode", "explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g", vmin=1, vmax=1, center=1, cmap="inferno", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap betwwen Variable")
heatmap.set_xticklables(heatmap.get_xticklables(), rotation=90)
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
# REGGERATION PLOT BETWEEN LOUDNESS & ENERGY

plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="c").set(title="Loudness vs Energy Correlation")
```

This shows -> as the loudness increases, the energy of the song also increases


```python
# REGGERATION PLOT BETWEEN POPULARITY & ACOUSTICNESS

plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="b").set(title="Popularity vs Acousticness Correlation")
```

This reggeration plot shows that when the acousticness increases, the popularity of the song decreases and vice versa


```python
# CREATE A NEW COLUMN "YEAR"

df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
# DISTRIBUTION PLOT - TO VISULATIZE TOTAL NO. OF SONGS EACH YEARSINCE 1922 THAT IS AVAILABLE ON SPOTIFY

sns.displot(years,discrete=True, aspect=2, height=5, kind="hist").set(title="Number of Songs per year")
```

```python
df_genre=pd.read_csv('SpotifyFeatures.csv')
```

```python
df_genre.head(5)
```

```python
# DURATION OF SONGS FOR DIFFERERNT GENRE
plt.title("Duration of the songs for different genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
# TOP 5 GENRE 
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("popularity", ascending=False).head(10)
sns.barplot(y="genre", x="popularity", data = famous).set(title="Top 5 Genre based on popularity")

```
